```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
plt.rc('font', family='NanumBarunGothic')
```

### 1. Line graph

```python
np.random.seed(2021)
df = pd.DataFrame(np.random.randn(100, 3),
                  index=pd.date_range('7/1/2021', periods=100),
                  columns=['A', 'B', 'C']).cumsum()
df.tail()
```

```python
df.plot()           # X-축: index, Y-축: A, B, C 컬럼, 범례: 컬럼명
plt.title("Pandas의 plot 메소드 사용 예")
plt.xlabel("날짜")
plt.ylabel("Data")
plt.grid()
plt.show()
```

### 2. Bar chart

```python
iris = sns.load_dataset('iris')

plt.figure(figsize=(12,6))
iris.sepal_length[:50].plot(kind='bar', rot=0)
plt.title('Sepal length of Setosa')
plt.show()
```

```python
df2 = iris.groupby('species').mean()
df2.columns.name = 'Feature'
df2
```

```python
df2.plot.bar(rot=0)
plt.title('각 종의 Feature별 평균')
plt.ylim(0,8.5)
plt.show()
```

### 3. Pie chart

```python
titanic = sns.load_dataset('titanic')
s = titanic['class'].value_counts()
s
```

```python
s.plot.pie(autopct='%.2f%%')
plt.title('선실별 승객수 비율')
#plt.axis('equal')
plt.show()
```

### 4. Box plot

```python
iris[iris.species == 'setosa'].plot.box()
plt.title('Setosa 품종의 각 Feature별 Box plot')
plt.show()
```
